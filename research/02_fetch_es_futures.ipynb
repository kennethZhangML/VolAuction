```python
import httpx 
import requests 
import yfinance as yf
import pandas as pd 
from datetime import datetime, timedelta
```

```python
spx = yf.Ticker("^GSPC")
spx_hist = spx.history(start="2025-01-28", end="2025-03-31", interval="1d")
```

```python
spx_hist = spx_hist[["Open", "Close", "High", "Low", "Volume"]].reset_index()
spx_hist.rename(columns={"Date": "date"}, inplace=True)
```

```python
spx_hist.to_csv("/Users/kennethzhang/Desktop/VolAuction/data/spx_spot_daily.csv", index=False)
```

```python
import time

start_date = datetime.strptime("2025-01-28", "%Y-%m-%d")
end_date = datetime.strptime("2025-03-31", "%Y-%m-%d")

all_data = []

current = start_date
while current <= end_date:
    print(f"Fetching {current.date()}...")
    es = yf.Ticker("ES=F")

    try:
        hist = es.history(period="7d", interval="1m")
        filtered = hist[hist.index.date == current.date()]
        filtered = filtered.copy()
        filtered["date"] = current.date()
        all_data.append(filtered)
    except Exception as e:
        print(f"Error on {current.date()}: {e}")
    
    current += timedelta(days=1)
    time.sleep(1)  
```

```python
df_all = pd.concat(all_data)
df_all.to_csv("/Users/kennethzhang/Desktop/VolAuction/data/es_futures_1m_all.csv")
print("Saved to es_futures_1m_all.csv")
```
